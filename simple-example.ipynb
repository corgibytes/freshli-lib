```
using XPlot.Plotly;

using Freshli;
using Freshli.Languages.Ruby;
using Freshli.Languages.Php;
using Freshli.Languages.Python;

ManifestFinder.RegisterAll();

FileHistoryFinder.Register<GitFileHistoryFinder>();

var runner = new Runner();
```

```
var projects = new Dictionary<string, IList<MetricsResult>>();
projects["pyspider"] = runner.Run("https://github.com/binux/pyspider");
projects["glances"] = runner.Run("https://github.com/nicolargo/glances");


var series = projects.Select(p => new Graph.Scattergl { 
  name = p.Key, 
  x = p.Value.Select(r => r.Date),
  y = p.Value.Select(r => r.LibYear.Total)
});


var chart = Chart.Plot(series.ToArray());
chart.WithTitle("LibYear over time");
display(chart)
```

```
var project = runner.Run("https://github.com/binux/pyspider");
var names = project.Select(r => r.LibYear.Select(p => p.Name));
```

```
var dates = project.Select(r => r.Date);
```

```
var totalLibYears = project.Select(r => r.LibYear.Total);
```

```
var names = project.Select(r => r.LibYear.Select(p => p.Name).ToArray()).Aggregate(
    new HashSet<string>(), 
    (hash, values) => {
        foreach(var value in values) {
            hash.Add(value);
        }
        return hash;
    }
);
names
```

```
var areaSeries = new List<Graph.Scattergl>();

foreach (var name  in names) {
    areaSeries.Add(new Graph.Scattergl {
        name = name,
        x = project.Select(r => r.Date),
        y = project.Where(r => r.LibYear.Any(p => p.Name == name)).Select(r => r.LibYear[name].Value)
    });
}

var chart = Chart.Plot(areaSeries.ToArray());
chart.WithTitle("LibYear over time per dependency");
display(chart)
```

```
var areaSeries = new List<Graph.Scatter>();

foreach (var name  in names) {
    areaSeries.Add(new Graph.Scatter {
        name = name,
        x = project.Select(r => r.Date),
        y = project.Where(r => r.LibYear.Any(p => p.Name == name)).Select(r => r.LibYear[name].Value),
        fill = "tonexty"
    });
}

var chart = Chart.Plot(areaSeries.ToArray());
chart.WithTitle("LibYear over time per dependency");
display(chart)
```
